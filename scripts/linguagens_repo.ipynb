## Paginando API

```python
load_dotenv()
access_token = os.getenv("ACESS_TOKEN")
headers = {'Authorization':'Bearer ' + access_token,
           'X-GitHub-Api-Version': '2022-11-28'}
```

```python
api_base_url = 'https://api.github.com'
owner = 'amzn' # username de quem vamos extrair os dados
url = f'{api_base_url}/users/{owner}/repos'
```

```python
repos_list = []
for page_num in range(1, 6):
    try:
        url_page = f'{url}?page={page_num}' # poderia passar o per_page => ?per_page=100
        response = requests.get(url_page, headers=headers)
        repos_list.append(response.json())
    except:
        repos_list.append(None)
```

```python
repos_list[0][2]['name']
```

```python
repos_name = []
repos_language = []

for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
        repos_language.append(repo['language'])
```

```python
dados_amz = pd.DataFrame()
dados_amz['repository_name'] = repos_name
dados_amz['language'] = repos_language
dados_amz.to_csv("amazon_github_repos.csv", index=False)
```

## Armazenando os dados

### Criando repositório com POST

```python
api_base_url = 'https://api.github.com'
url = f'{api_base_url}/user/repos'
url
```

```python
data = {
    'name': 'linguages_repositorios',
    'description': 'Linguagens de repositorios',
    'private': True
}

response = requests.post(url, json=data, headers=headers)
response.status_code
```

### Formato do arquivo

```python
import base64
with open('amazon_github_repos.csv', 'rb') as file:
    file_content = file.read()

encoded_content = base64.b64encode(file_content)
```

### Upload de arquivo com PUT

```python
api_base_url = 'https://api.github.com'
username = 'renanortegax'
repo = 'linguages_repositorios'
path = 'amazon_github_repos.csv'

url = f'{api_base_url}/repos/{username}/{repo}/contents/{path}'
url
```

```python
data = {
    'message': 'Initial commit - adicionando um arquivo',
    'content': encoded_content.decode('utf-8')
}

response = requests.put(url, json=data, headers=headers)
response.status_code
```
